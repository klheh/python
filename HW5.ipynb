#[2번 문제]

```python
class Ironman:
    def jarvis(self):
        print "The truth is.... I am Ironman"
        
class Mark1(Ironman):
    def jarvis(self):
        print "Mark1"
        
class Mark7(Ironman):
    def jarvis(self):
        print 'Mark7'
        
class Mark42(Ironman):
    pass

for marks in (Mark1(), Mark7(), Mark42()):
    marks.jarvis()
```

* 위의 예제에서 Ironman에 있는 jarvis를 Mark1, Mark4가 override합니다.
* Mark42에는 jarvis가 없으므로, Ironman의 내용을 그대로 사용하게 됩니다.
* for반복문에서 in안에 3개의 객체가 들어있는데, marks의 객체형이 동적으로 결정됩니다. 따라서 그때마다 jarvis가 각각의 객체에 맞게 다르게 사용됩니다.

요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

        >>> c = c + 5 
        >>> d = d - 5 
        >>> print c 
        [Count (step: 1)] 17 
        >>> print d 
        [Count (step: 2)] 9 

    => 위 요구사항에 대해 클래스 내에 __add__(self, other)와 def __sub__(self, other)를 정의하였습니다.
    각 내부는 return Counter(self.count + other, self.step), return Counter(self.count - other, self.step)와 같은데, +연산을 하였을 경우, 인스턴스(self)의 count값과 연산되는 값을 더한것(혹은 뺀 것)을 새로 만들어지는 인스턴스의 count값으로, 그리고 기존의 step을 이용하면 제시된 두 개의 산술 연산이 수행됩니다.

요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

    >>> print c > 10 
    True 
    >>> print d > 10 
    False 
    >>> print c < 10 
    False 
    >>> print d < 10 
    True 
    >>> print c == 17 
    True 
    >>> print d != 9 
    False 

    => 관계 연산을 위해 __gt__(self, other), __lt__(self, other), __eq__(self, other), __ne__(self, other)를 정의해주었습니다.
    함수 내용은 return self.count > other과 같은데, T/F가 출력되므로 반환값은 비교한 것을 반환해주면 됩니다. 각 연산에 맞게 >, <, ==, !=를 넣어줌으로써 수행될 수 있도록 하였습니다.

```python
class Counter:
    def __init__(self, cnt, stp=1):
        self.count = cnt
        self.step = stp
        
    def __str__(self):
        return '[Count (step : %s)] %s' % (self.step, self.count)
    
    def incr(self):
        self.count = self.count + self.step
        
    def __call__(self):
        self.incr()
        
    def __add__(self, other):
        return Counter(self.count + other, self.step)
    def __sub__(self, other):
        return Counter(self.count - other, self.step)
    
    def __gt__(self, other):
        return self.count > other
    def __lt__(self, other):
        return self.count < other
    def __eq__(self, other):
        return self.count == other
    def __ne__(self, other):
        return self.count != other
        
c = Counter(10)
d = Counter(10, 2)

print c
print d
print
        
c.incr()
d.incr()
print c
print d
print

c()
d()
print c
print d
print

c = c + 5 
d = d - 5 
print c
print d
print

print c > 10
print d > 10
print c < 10
print d < 10
print c == 17
print d != 9
```

#[4번 문제]
다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)                     
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

* eliminate_duplicate는 정적 메소드로 인스턴스 객체와 무관하게 클래스 이름 공간에 존재하는 메소드로서 클래스 이름을 이용하여 직접 호출할 수 있는 메소드입니다. 장식자 @staticmethod 를 통해 정적 메소드임을 알 수 있습니다. 위 메소드는 인자로 받은 리스트 l를 그 안에서 중복된 원소를 뺀 리스트로 다시 만들어 주는 역할을 하는 메소드입니다. 처음 반복문에서 l에서 중복되지 않은 원소들을 e로 뽑아내고, l을 비운 다음, e를 다시 l에 넣는 두번째 반복문을 볼 수 있습니다.
* __init__(self, l)를 통해 인자로 들어온 l의 원소를 갖는 인스턴스를 append()함수를 이용하여 만들고, eliminate_duplicate()함수를 이용하여 중복된 원소를 제거해줍니다.
* __str__(self)를 통해 출력시 적절한 모양(?)으로 만들어 줍니다. 반복문을 이용해서 result에 '원소값, '를 더해주고, 반복문이 끝난 후 마지막 원소 뒤의 ' ,'를 빼고 '}'를 넣음으로써 적절한 형태가 완성됩니다. 따라서 출력할 경우, MySet: {1 ,2 ,3} 과 같은 형태로 출력됩니다.

정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.

    >>> u = s | t

    => 클래스 내에 __or__(self, other)를 정의하여, 요구사항을 수행합니다. 위의 예에서 self가 s, other가 t라고 생각하면 됩니다. __init__와 유사한 방법으로, 우선 temp에 self와 other의 원소를 추가한 후 eliminate_duplicate(self)를 이용하여 중복된 원소를 제거하면 합집합이 되며, 그 합집합을 반환하면 됩니다.

요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

    >>> u = s & t 
    >>> print u 
    MySet: {2, 3} 

    => 클래스 내에 __and__(setf, other)를 정의하여 요구사항을 수행합니다. sumset이라는 함수를 따로 만들어서 두 인스턴스의 원소중 겹치는 것을 뽑아내는 과정을 수행합니다. set1을 수행한 결과값을반환하면 됩니다. set1은 eliminate_duplicate와 같은 정적 메소드로, self의 원소를 반복하는데, other에 들어있는 원소 중 같은것이 있다면 s에 추가합니다. l(= self)를 비운 후, s의 원소를 저장하고 반환해줍니다.

요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

    >>> s = MySet([1, 2, 3]) 
    >>> t = MySet([3, 4, 5]) 
    >>> u = s - t 
    >>> print u 
    MySet: {1} 

####=> 차집합인 s-t일 경우, s의 원소 중 s와 t의 교집합의 원소에 들어가는 것을 뺀 나머지라 {1, 2}가 나와야 한다고 생각하기에 그런 방식으로 풀었습니다. 

=> 클래스 내에 __sub__(self, other)를 정의하여 요구사항을 수행합니다. 우선 temp1에는 self의 원소를 넣고, temp2에는 앞에 만들었던 set1함수를 이용하여 교집합을 저장합니다. 그 다음 set2함수의 결과를 반환합니다. set2함수는 정적 메소드로, 빈 리스트 s에 만약 temp1의 원소가 temp2의 원소에 없다면 s에 추가합니다. s는 차집합이고 그것을 MySet2(s)로 반환하면 원하는 답이 반환됩니다. 

```python
class MySet2(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet2.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet2: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
    
    def __or__(self, other):
        temp = []
        for e in self:
            temp.append(e)        
        for h in other:
            temp.append(h)
        MySet2.eliminate_duplicate(temp)
        return MySet2(temp)
    
    def __and__(self, other):
        return MySet2.set1(self, other)
    
    def __sub__(self, other):
        temp = []
        for e in self:
            temp.append(e)  
        temp2 = MySet2.set1(self, other)
        return MySet2.set2(temp, temp2)
        
        

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)  
            
    @staticmethod    
    def set1(l, k):
        s = []
        for e in l:
            if e in k:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e) 
        return l
            
    @staticmethod    
    def set2(t1, t2):
        s = []
        for e in t1:
            if e not in t2:
                s.append(e)
        return MySet2(s)
    
if __name__ == "__main__":
    s = MySet2([1, 2, 2, 3])
    print s
    t = MySet2([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
    u = s | t
    print u 
    u = s & t
    print u 
    
    s = MySet2([1, 2, 3])
    t = MySet2([3, 4, 5]) 
    u = s - t
    print u 
```

참고: 빈 칸이나 하이픈('-')은 셈에서 제외하며, 단어 사이의 and 는 셈에 넣습니다.

#[11번 문제]
###프로젝트오일러 20번
n! 이라는 표기법은 n × (n − 1) × ... × 3 × 2 × 1을 뜻합니다.

예를 들자면 10! = 10 × 9 × ... × 3 × 2 × 1 = 3628800 이 되는데,
여기서 10!의 각 자리수를 더해 보면 3 + 6 + 2 + 8 + 8 + 0 + 0 = 27 입니다.

100! 의 자리수를 모두 더하면 얼마입니까?

```python
#팩토리얼 함수
def fac(k):
    n = 1
    for i in range(1, k + 1):
        n *= i
    return n

res = 0
num = str(fac(100))

#자리수 더하기
for i in num:
    res += int(i)

print res
```

=> 팩토리얼 함수를 정의해 놓은 후(이전 과제 참고), 100!을 구한 값을 문자열로 받고, 그 문자열에서 원소들을 반복하는데, 원소는 각 자리수가 됩니다. 따라서 그 값들을 숫자로 받아와 더하면 문제의 정답이 됩니다.

n의 약수들 중에서 자신을 제외한 것의 합을 d(n)으로 정의했을 때,
서로 다른 두 정수 a, b에 대하여 d(a) = b 이고 d(b) = a 이면 
a, b는 친화쌍이라 하고 a와 b를 각각 친화수(우애수)라고 합니다.

예를 들어 220의 약수는 자신을 제외하면 1, 2, 4, 5, 10, 11, 20, 22, 44, 55, 110 이므로 그 합은 d(220) = 284 입니다.
또 284의 약수는 자신을 제외하면 1, 2, 4, 71, 142 이므로 d(284) = 220 입니다.
따라서 220과 284는 친화쌍이 됩니다.

10000 이하의 친화수들을 모두 찾아서 그 합을 구하세요.

```python
#약수 합 구하기
def f1(a):
    n = 1
    temp = []
    while(n < a):
        if(a % n == 0):
            temp.append(n)
        n += 1
    return sum(temp)

result = []

#친화수의 합 구하기
for a in range(2, 10001):
    b = f1(a)
    n2 = f1(b)
    if(a == n2 and a != b):
        result.append(a)
        result.append(b)
result = list(set(result))
print result
print sum(result)
```

=> 우선 f1은 자신을 제외한 약수의 합을 구하는 함수입니다. 1부터 입력된 수보다 작을때 까지 반복하는데, 나누었을때 0이라면 temp 리스트에 추가하고 , 반복문이 끝났을 때, temp리스트의 합입니다.
그 다음, 2부터 10000까지 반복하는데, f1(a)의 값이 a의 약수중 자신을 제외한 약수의 합이고, 그 합을 다시 f1(b)했을때 나오는 값이 a와 같다면 a와 b는 친화쌍일 것 입니다. 단, a와b가 같은 수일 경우가 있으므로 그 부분을 조건문을 이용하여 제외합니다. 
중복된 수가 들어갈 가능성이 있으므로 set()과 list()를 이용하여 제거해주고 그 리스트의 합이 문제의 정답입니다.

```python
import urllib2
import string
import operator

class WebWordsFrequency:
    # 요구사항2번) 생성자에 가변 인수 설정
    def __init__(self, *s):
        self.urllist = list(s) # url 리스트
        self.cnt = len(list(s)) # url 보유 숫자
        
    #요구사항7번) 가장 많이 출현한 단어 리스트 반환    
    def getMaxFreqencyWords(self):
        self.getWordsFrequency() # 단어 사전을 위해 호출
        if self.cnt == 0: # 등록된 URL없는경우
            return None
        else :
            self.resultlist = []
            self.sorted_dic = sorted(self.dic.iteritems(), key=operator.itemgetter(1)) #사전 크기순으로 정렬
            for q in range(1, self.cnt):
                if(self.sorted_dic[-1][1] == self.sorted_dic[-q][1]): #숫자 같은지 확인
                    self.resultlist.append(self.sorted_dic[-q][0])  #같을 경우 결과 리스트에 추가
                else:
                    break
            return self.resultlist 
    
    #요구사항6번) 단어 출현 빈도 사전 - 이전 과제 참고
    def getWordsFrequency(self):
        self.dic = dict() #사전 생성
        for k in range(self.cnt):
            source = urllib2.urlopen(self.urllist[k]).read()

            for i in range(0, source.count('<!--')):
                if source.find("<!--") != -1:
                    source = source.replace(source[source.find("<!--"):source.find("-->")+3:],"")

            for i in range(0,source.count("<")):
                if source.find("<") != -1:
                    source = source.replace(source[source.find("<"):source.find(">")+1:],"")
                    
            words = source.split()
            s = ' '.join(words)
            
            #구두문자 제거
            words = list(s)
            for i in range(len(words)):    
                for k in range(len(string.punctuation)):
                    if(words[i] == string.punctuation[k]):
                        words[i] = ' '
            words = ''.join(words)

            #구두문자가 제거된 문자열을 공백을 제외한 리스트로 만들기
            words = words.split()
            
            #영어 불용어 제거
            eng = ['a', 'and', 'are', 'as', 'at', 'be', 'but', 'by', 'for', 'if', 'in', 'into', 'is', 'it', 'no', 'not', 'of', 'on', 'or', 's', 'such', 't', 'that', 'the', 'their', 'then', 'there', 'these', 'they', 'this', 'to', 'was', 'will', 'with']
            for i in range(len(words)):
                for k in range(len(eng)):
                    if words[i] == eng[k]:
                        words[i] = ' '
            words = ' '.join(words)

            #영어 불용어가 제거된 문자열을 공백을 제외한 리스트로 만들기
            words = words.split()
            
            #사전 만들고 카운트 세기
            for word in words:
                if word not in self.dic:
                    self.dic[word] = 1
                else:
                    self.dic[word] += 1                  
            return self.dic        
        
    # 요구사항3번) URL을 추가
    def addUrl(self, au):
        self.urllist.append(au) #url 리스트에 추가
        self.cnt += 1 #url 보유 숫자 증가

    # 요구사항4번) URL을 삭제
    def removeUrl(self, ru):
        check = 0
        #삭제할 url이 있는지 확인
        for m in range(self.cnt):
            if(self.urllist[m] == ru):
                check = 1
                break
            
        #삭제하고자 하는 url이 등록되어있지 않는 경우
        if (check == 0):
            print "Not found"
            return
        
        #있는 경우
        ind = self.urllist.index(ru) #인덱스값 찾기
        self.urllist.pop(ind) #제거
        self.cnt -= 1 #url 보유 숫자 감소
      
    # 요구사항5번) URL을 출력
    def listUrls(self):
        for k in range(self.cnt):
            print self.urllist[k]
            
        # 보유한 url이 없는 경우
        if (self.cnt == 0):
            print "No Url"
        
w1 = WebWordsFrequency('http://www.cnn.com', 'http://www.times.com', 'https://www.amazon.com') 
w2 = WebWordsFrequency('http://www.cnn.com', 'http://www.times.com') 
w3 = WebWordsFrequency()

w1.addUrl('https://github.com')
w3.addUrl('http://stackoverflow.com')

w1.removeUrl('http://www.cnn.com')
w3.removeUrl('http://stackoverflow.com')
w3.removeUrl('http://www.cnn.com')
print
w1.listUrls()
print
w2.listUrls()
print
w3.listUrls()
print
print
print w1.getWordsFrequency() 
print
print
print w1.getMaxFreqencyWords()
print w2.getMaxFreqencyWords() 
print w3.getMaxFreqencyWords()
```

```python
class OrderedWebWordsFrequency(WebWordsFrequency):
    def __init__(self, *s):
        self.urllist = list(s) # url 리스트
        self.cnt = len(list(s)) # url 보유 숫자
        
    def getWordsFrequency(self, reverse = False):
        #이 부분은 이전 과제와 거의 유사, 정렬 부분에 변화 O
        self.dic = dict() #사전 생성
        for k in range(self.cnt):
            source = urllib2.urlopen(self.urllist[k]).read()

            for i in range(0, source.count('<!--')):
                if source.find("<!--") != -1:
                    source = source.replace(source[source.find("<!--"):source.find("-->")+3:],"")

            for i in range(0,source.count("<")):
                if source.find("<") != -1:
                    source = source.replace(source[source.find("<"):source.find(">")+1:],"")
                    
            words = source.split()
            s = ' '.join(words)
            
            #구두문자 제거
            words = list(s)
            for i in range(len(words)):    
                for k in range(len(string.punctuation)):
                    if(words[i] == string.punctuation[k]):
                        words[i] = ' '
            words = ''.join(words)

            #구두문자가 제거된 문자열을 공백을 제외한 리스트로 만들기
            words = words.split()
            
            #영어 불용어 제거
            eng = ['a', 'and', 'are', 'as', 'at', 'be', 'but', 'by', 'for', 'if', 'in', 'into', 'is', 'it', 'no', 'not', 'of', 'on', 'or', 's', 'such', 't', 'that', 'the', 'their', 'then', 'there', 'these', 'they', 'this', 'to', 'was', 'will', 'with']
            for i in range(len(words)):
                for k in range(len(eng)):
                    if words[i] == eng[k]:
                        words[i] = ' '
            words = ' '.join(words)

            #영어 불용어가 제거된 문자열을 공백을 제외한 리스트로 만들기
            words = words.split()
            
            #사전 만들고 카운트 세기
            for word in words:
                if word not in self.dic:
                    self.dic[word] = 1
                else:
                    self.dic[word] += 1 
                    
            # 정렬하기 -> 변화 O 부분
            if (reverse == False): #내림차순!
                self.sortlist = sorted(self.dic.iteritems(), key=operator.itemgetter(1), reverse=True) #역순이므로 reverse를 True로 바꾸어 줌.
            else: #오름차순!
                self.sortlist = sorted(self.dic.iteritems(), key=operator.itemgetter(1))# 역순이 아니므로 default값 False
            return self.sortlist
        

w4 = OrderedWebWordsFrequency('http://www.times.com', 'https://www.amazon.com', 'https://github.com') 
print w4.getWordsFrequency()
print
print 
print w4.getWordsFrequency(reverse=True)
```
