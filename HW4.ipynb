1) 파일 s.txt을 읽어서 각 라인에 있는 첫 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s1.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s1.txt에 있는 파일 내용은 다음과 같다.

```python
f = open('s.txt')
i = 1
k = []
for line in f:
    k += line,
    i += 1
s = ''.join(k)
k = s.split('\n')
k.sort()
f.close()

f = open('s1.txt', 'w')
f.write('\n'.join(k))
f.close()

f = open('s1.txt')
f.seek(0)
print f.read()
f.close()
```

* 파일을 읽은 후, 반복문을 통해 k에 라인 별로 넣습니다.(리스트)
* k에는 \n이 들어있기 때문에 \n을 제거해주기 위해, ''.join(k)를 해주고, split('\n')을 해주면, ['pig ham', 'cat dog', 'ham bird', 'dog pig']이 됩니다.
* sort()를 이용하여 k를 정렬해주고, s파일은 닫고, s1파일을 엽니다.기록할것이기 때문에 w로 열어줍니다. f.write()를 통해 파일에 적어주는데, k를 마지막에 '\n'을 이용하여 연결한 결과를 저장한다면 각 원소들이 한 줄씩 적혀집니다.

2) 파일 s.txt을 읽어서 각 라인에 있는 두 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s2.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s2.txt에 있는 파일 내용은 다음과 같다.

```python
def cmp_1(a1, a2):
    n1 = a1.find(' ')
    n2 = a2.find(' ')
    return cmp(a1[n1+1], a2[n2+1])

f = open('s.txt')
i = 1
k = []
for line in f:
    k += line,
    i += 1
k = ''.join(k).split('\n')
k.sort(cmp_1)
f.close()

f = open('s2.txt', 'w')
f.write('\n'.join(k))
f.close()

f = open('s2.txt')
f.seek(0)
print f.read()
f.close()
```

* cmp_1함수는 두번째단어는 공백 뒤에 오므로 우선 공백의 위치를 찾은 다음, 그 다음의 문자열을 비교하는 함수입니다.

###[3번 문제]
range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.

    range(stop)
    
    range(start, stop[, step])

2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.

3) frange 사용 예

    3-1) frange(0.5)

        [0.0, 0.1, 0.2, 0.3, 0.4]

    3-2) frange(1.0, 2.0)

        [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]

    3-3) frange(2.2, 4.0, 0.5)

        [2.2, 2.7, 3.2, 3.7]   

```python
def f1(stop):
    #1개 들어오는경우(stop만 있는경우)
    stp = float(stop[0])
    i = 0.0
    k = []
    while i < stp:
        k.append(i)
        a = 0.1
        i += a
        i = round(i, 2)
    return k

def f2(stop):
    #2개나 3개 들어오는 경우
    if(len(stop) == 2):
        start = float(stop[0])
        stp = float(stop[1])
        step = 0.1
    elif(len(stop) == 3):
        start = float(stop[0])
        stp = float(stop[1])
        step = float(stop[2])
    else:
        print "Input Error"
        return
    
    
    k = []
    i = start
    while i < stp:
        k.append(i)
        i += step
        i = round(i, 2)
    return k
    
    

def frange(*a):
    s = str(a)[1:-1].split(',')
    if(s[-1] == ''):
        s.pop(-1)
        
    if(len(s) == 1):
        return f1(s)
    else:
        return f2(s)

print frange(1)
print frange(1,2)
print frange(0.5)
print frange(1.0, 2.0)
print frange(2.2, 4.0, 0.5)
```

* 함수의 인자가 다양하게 들어올 수 있으므로, def frange(*a) 와 같이 인자를 정한 후, 인자의 수가 1개라면 f1함수를 실행하고 2개 이상이라면 f2함수를 실행합니다.
* 1개가 들어오는 경우, stop값만 있으므로 나머지는 문제에서 주어진 초기값을 이용하여 빈 리스트k에 start 값부터 stop값까지 step만큼 더하면서 추가하는 것을 반복하여 최종 결과값을 반환합니다.
* 2개, 3개가 들어오는 경우 또한 마찬가지로 들어오지 않은 값만 정해주고 위와 같은 방법으로 함으로써 해결합니다.

###[4번 문제]
가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오

    sum() 사용 예
        sum()
            0
        sum(1, 2)
            3
        sum(1, 2, 3, 4, 5)
            15
        sum(1, 5, 7, 2, -10)
            5

```python
def sum1(*a):
    s = str(a)[1:-1].split(',')
    result = 0
    if(s[-1] == ''):
        s.pop(-1)
        
    if(len(s) == 0):
        return 0
    else:
        for i in s:
            result += int(i)
        return result
    

    
print sum1()
print sum1(1, 2)
print sum1(1, 2, 3, 4, 5)
print sum1(1, 5, 7, 2, -10)
```

* 아래 문제에서 sum함수를 사용하기 때문에 sum1으로 함수명을 정하였습니다.
* 이전 문제와 유사하게 인자가 여러개 들어오는것을 처리해주기 위해 def sum1(*a)의 방식으로 함수를 정의하며, ','문자를 이용하여 입력받은 값들을 나누고 문자열이므로 int()함수를 이용하여 정수로 변환한 후, result에 값을 더해줌으로써 최종적으로 result를 구합니다.

###[5번 문제]

    1) 다음에 제시되는 함수들을 모두 이용해야 한다.
        split
        map
        join
    2) myinitial() 함수 사용 예
        myinitial("as soon as possible")
            ASAP

```python
def fir(s):
    return s[0]
    
def myinitial(sentence):
    L = sentence.split(' ')
    L = map(fir, L)
    S = ''.join(L)
    S = S.upper()
    return S
    
print myinitial("as soon as possible")
```

* myinitial에서 입력받은 문장을 split을 이용하여 각 단어로 나눈 다음, map함수를 이용하여 각 원소마다 fir함수를 이용하여 처리합니다. fir함수는 각 문자열의 첫번째 원소만 반환하는 함수입니다. 
* 그 결과를 join함수로 연결하고, upper함수를 이용하면 대문자들만 출력하게됩니다.

###[6번 문제]
음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

[옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.

```python
import sys
sys.setrecursionlimit(10000)

def myfac(n):
    if(n <= -1):
        return "Input Error"
    else:
        if(n<=1):
            return 1
        
        return n * myfac(n-1)
    
print myfac(-1)
print myfac(0)
print myfac(100)
```

* 팩토리얼은 5! = 5 * 4!, 4! = 4 * 3! ....이므로
* 팩토리얼을 구하는 함수를 재귀적으로 구현하면 0과 1일 경우는 1을 반환하고, 그보다 클 경우는 n * myfac(n-1)을 하면 됩니다.
* 그대로 1000까지 진행하게 되는 경우, 에러가 발생하므로 import sys sys.setrecursionlimit(10000)를 추가하였습니다,

###[8번 문제]
(프로젝트오일러 10번)
10 이하의 소수를 모두 더하면 2 + 3 + 5 + 7 = 17 이 됩니다.

이백만(2,000,000) 이하 소수의 합은 얼마입니까?

```python
def primes(n):
    p = {}
    q = []
    result = 0
    # 2부터 N까지 1인 사전만들기
    for i in range(1, n):
        p[i+1] = 1
        
    #사전 key(2~N) 반복
    for i in p.keys():        
        if p[i] == 1:
            #배수 제거
            for k in range(i * i, n + 1, i):
                p[k] = 2

    #1로 남은 애들 찾기
    for i in p.keys():
        if p[i] == 1:
            q.append(i)
            
    #합구하기
    for i in q:
        result += i
    return result

print primes(2000000)
```

* 에라토스네의 체 : 에라토스테네스가 고안해낸 주요 처리절차는 다음과 같습니다.
    2부터 N까지를 크기로 하는 체를 만듭니다.
    우선 모든 체의 요소를 1로 채웁니다.
    2를 제외한 2의 배수를 체에서 모두 제거(즉, 해당 배열 요소들을 모두 2으로 설정)합니다.
    3을 제외한 3의 배수를 ... (위와 동일)
    위의 절차를 N까지 반복해서 최종적으로 1을 가지고있는 숫자가 바로 소수입니다.
* 위 방법을 체를 사전으로 생각하여 빈 사전에 2부터 N까지의 사전을 만들고 1을 입력합니다. 2~N까지 반복하면서 해당 값이 1이면, 그 수의 배수의 값을 2로 만듦으로써 제거함을 반복하고, 1인 값만 q에 추가하게 되면 q에는 소수가 저장되며, 그 합을 구하면 결과값이 나옵니다.

###[10번 문제]
(프로젝트오일러 12번)1부터 n까지의 자연수를 차례로 더하여 구해진 값을 삼각수라고 합니다.
예를 들어 7번째 삼각수는 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28이 됩니다.
이런 식으로 삼각수를 구해 나가면 다음과 같습니다.

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...
이 삼각수들의 약수를 구해봅시다.

     1: 1
     3: 1, 3
     6: 1, 2, 3, 6
    10: 1, 2, 5, 10
    15: 1, 3, 5, 15
    21: 1, 3, 7, 21
    28: 1, 2, 4, 7, 14, 28
위에서 보듯이, 5개 이상의 약수를 갖는 첫번째 삼각수는 28입니다.

그러면 500개 이상의 약수를 갖는 가장 작은 삼각수는 얼마입니까?

```python
n1 = 1
n2 = 2 

def f3(n):
    c = 0
    for i in range(int(n**(0.5))):
        if n % (i + 1)==0:
            c += 1
    return c

while 1:
    if f3(n1) > 250:
        break
    else:
        n1 += n2
        n2 += 1
print n1
```

* 삼각수이므로, 1부터 시작하여 2, 3, 4... 를 차례대로 더하기 때문에 n1는 기준 값이고, n2는 더해질 값 입니다. 기준값은 기존의 기준값에 더해질 값을 더하는 것이고, 더해질 값은 1씩 증가합니다.
* 자연수 n이 n = a * b를 만족한다면 a, b 는 모두 n의 약수이므로 제곱근까지만 검사하면 됩니다. 따라서 약수의 개수를 검사하는 함수 f3에서 범위를 n의 제곱근인 n**(0.5)로 하였습니다.
* f3은 약수의 개수를 반환해주는 함수인데, n을 1~n의 제곱근까지 나누었을때 0이면 약수이므로 약수의 개수인 c를 1 증가시켜줍니다. range(n)는 0부터 n-1까지이므로 i에 1을 더해줍니다.
* 위에서 말했듯이, 자연수 n이 n = a * b를 만족한다면 a, b 는 모두 n의 약수이므로 500개 이상의 약수를 갖는 경우는, 범위를 제곱근까지만 검사하였을 때, 약수의 개수가 250개보다 큰 경우입니다. 따라서 만약에 f3(n)의 결과가 250보다 크다면 n이 500개 이상의 약수를 갖는 삼각수이며, 처음 나오는 경우가 가장 작은 경우이므로 break를 통해 반복문을 빠져나오고 그 값이 결과값 입니다.

###[12번 문제]
(프로젝트오일러 14번) 양의 정수 n에 대하여, 다음과 같은 계산 과정을 반복하기로 합니다.

n → n / 2 (n이 짝수일 때)
n → 3 n + 1 (n이 홀수일 때)

13에 대하여 위의 규칙을 적용해보면 아래처럼 10번의 과정을 통해 1이 됩니다.

13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1
아직 증명은 되지 않았지만, 이런 과정을 거치면 어떤 수로 시작해도 마지막에는 1로 끝나리라 생각됩니다. 
(역주: 이것은 콜라츠 추측 Collatz Conjecture이라고 하며, 이런 수들을 우박수 hailstone sequence라 부르기도 합니다)

그러면, 백만(1,000,000) 이하의 수로 시작했을 때 1까지 도달하는데 가장 긴 과정을 거치는 숫자는 얼마입니까?

참고: 계산 과정 도중에는 숫자가 백만을 넘어가도 괜찮습니다.

```python
n = 0
m = 0

for i in range(3, 1000000, 2):
    t = i
    c = 0
    
    #반복
    while i != 1:
        if i % 2 == 0:
            i = i/2 
        else:
            i = i * 3 + 1
        c += 1
        
    #최대값
    if c > m:
        m = c
        n = t
print n
```

* 우선, 홀수가 짝수보다 카운트가 더 많게되므로 홀수에 대해서만 테스트합니다. 1은 관계없으므로 3부터 시작합니다. 따라서 반복문의 범위는 range(3, 1000000, 2)가 됩니다.
* 반복을 i가 1이 아닐때 하는데(즉, 1이 되기 전까지 반복) n → n / 2 (n이 짝수일 때 -> i % 2 == 0), n → 3 n + 1 (n이 홀수일 때)를 수행합니다.

###[13번 문제]
(프로젝트오일러 15번) 아래와 같은 2 × 2 격자의 왼쪽 위 모서리에서 출발하여 오른쪽 아래 모서리까지 도달하는 길은 모두 6가지가 있습니다 (거슬러 가지는 않기로 합니다).


그러면 20 × 20 격자에는 모두 몇 개의 경로가 있습니까?

```python
def f4(n):
    m = 1
    for i in range(1, n + 1):
        m = m * i
    return m

print f4(40) / (f4(20) * f4(20))
```

* 위 문제는 경우의 수에 관련된 문제로, 조합으로 생각하면 간단합니다. 2x2의 경우 4개중 2개를 선택하는 조합의 수와 동일합니다. 문제는 20x20이므로 40C20을 구하는 문제입니다.
* 우선, f4는 팩토리얼 함수 입니다. 1부터 n까지 반복해야하므로 range(1, n+1)입니다. 팩토리얼은 5! = 5 4!, 4! = 4 3! ....이므로, 1부터 n까지 반복하면서 기존까지의 수에 현재의 수를 곱하면됩니다.
* 문제에서 구하는 40C20은 40!/(20! * 20!)이므로, f4(40) / (f4(20) * f4(20))를 하면 결과값이 나옵니다.
